We have a problem with dataset which is the curse of imbalance dataset. it prevents any machine learning model from learning the properties of the class with less data than the majority class. 
I managed to handle this problem by applying SMOTE algorithm to generate data for our stroke class and this balanced our stroke class. 
Now we have a balanced dataset in terms of our target with equal values.

```python
print ("our target which is stroke BEFORE balancing : " )
print (y.value_counts())
```

```python
smote = SMOTE (random_state = 42)
x_balanced , y_balanced = smote.fit_resample (x , y)
```

```python
x_train , x_test , y_train , y_test = train_test_split (x  , y , test_size = 0.3 , random_state = 42 , stratify = y)
```

```python
random_forest = RandomForestClassifier (n_estimators = 100 , random_state = 42  )
random_forest.fit (x_train , y_train)

y_pred = random_forest.predict (x_test)
```

```python
print ("Random forest validation")
print ("Accuracy : {:.2f}".format (accuracy_score (y_test,y_pred)))
print ("percision : {:.2f}".format (precision_score (y_test , y_pred)))
print ("Recall : {:.2f}".format (recall_score (y_test , y_pred)) )
print ("f1 score {:.2f}".format (f1_score (y_test , y_pred)))
```

```python
print ("classification report : \n " , classification_report (y_test , y_pred))
```

Here i have evaluated my performance with a new patient data. I have managed to create a definition for the new patient's data which receives required data in the correct range and if it is not in the correct range, it will show an error based on the incorrect value we have given to the algorithm. Patient's data will be transformed as a data frame and then using the RandomForest algorithm, we can predict whether this new patient has the risk of stroke (1) or not (0). There is also a percentage of the risk which can be usefull as well. 

```python
def predict_stroke(patient_data):
    valid_ranges = {
        'gender': [1, 2],
        'age': [1, 2, 3],
        'hypertension': [0, 1],
        'heart_disease': [0, 1],
        'avg_glucose_level': [1, 2, 3],
        'bmi': [1, 2, 3],
        'smoking_status': [1, 2, 3]
    }
    
    for key, value in patient_data.items():
        if value not in valid_ranges[key]:
            raise ValueError(f"Invalid value for {key}: {value}. Must be one of {valid_ranges[key]}")
    
    patient_df = pd.DataFrame([patient_data])
    

    prediction = random_forest.predict(patient_df)
    probability = random_forest.predict_proba(patient_df)[0][1]
    
    return {
        'stroke_prediction': 'Yes' if prediction[0] == 1 else 'No',
        'stroke_probability': probability
    }

# Example patient
new_patient = {
    'gender': 2,  
    'age': 3,     
    'hypertension': 0,  
    'heart_disease':0 ,  
    'avg_glucose_level': 3,  
    'bmi': 2,     
    'smoking_status': 3
}

try:
    result = predict_stroke(new_patient)
    print("\nNew Patient Prediction:")
    print(f"Stroke Risk : {result['stroke_prediction']}")
    print(f"Probability of Stroke : {result['stroke_probability']:.2f}")
except ValueError as e:
    print(f"Error: {e}")
```

There are some features in this method which are more important than others. For example : 1.hypertension , 2.heart_disease 3.smoking_status are the exmaples of this idea. So we can observe that whenever a new patient has higher values in these features (higher smoking status as an example which says a smoker or regular smoker) can lead to higher probability of stroke and stroke risk.
